# 1 - Obtendo os dados geográficos

### Seleciona em 'indicadores.db' apenas o df 'cities'

```python
cnx = sqlite3.connect('indicadores.db')
cities_df = pd.read_sql_query("SELECT * FROM cities_df", cnx)
cities_df
```

### Seleciona em 'municipios.csv' apenas os atributos que vamos trabalhar

```python
df_geobr = pd.read_csv('municipios.csv')
df_geobr = df_geobr[['codigo_ibge','latitude','longitude']]
df_geobr
```

### Merge para adicionar os dados de latitude e longitude de cada município em 'cities_df'

```python
cities_df.reset_index(inplace=True)
cities_df = cities_df.merge(df_geobr,how='inner', left_on='ibge_code', right_on='codigo_ibge')
cities_df
```

### Formatando data:

```python
# Setting the right format for starttime and stoptime
cities_df['week_last_day'] = cities_df['week_last_day'].str[:10]
cities_df['week_last_day'] = pd.to_datetime(cities_df['week_last_day'])

cities_df
```

### Drop nas colunas extras:

```python
cities_df.drop('index', axis='columns', inplace=True)
cities_df.drop('last_available_confirmed', axis='columns', inplace=True)
cities_df.drop('last_available_deaths', axis='columns', inplace=True)
cities_df.drop('prevalence', axis='columns', inplace=True)
cities_df.drop('mortality', axis='columns', inplace=True)
cities_df.drop('lethality', axis='columns', inplace=True)
cities_df.drop('codigo_ibge', axis='columns', inplace=True)

cities_df
```

### Fronteiras dos estados:

opacity  --> valores entre 0 e 1 definem a transparência da camada (com 0 ela fica invisível)

weight   --> define a grossura das fronteiras

color     --> define a cor das fronteiras em hexadecimal

fillColor --> define a cor de preenchimento em hexadecimal


```python
shape_br = read_state(code_state='all')
shape_br = shape_br.rename(columns={'abbrev_state':'state'})

style = {'opacity': 0.7, 'weight': 1,'fillColor': '#ffffff', 'color': '#000000'}
```

```python
data_city = cities_df[['week_last_day',indicador,'latitude','longitude','region','state','total_inhabitants','epidemiological_week','city']]
data_city
```

### Removendo linhas sem dados confirmados de COVID-19:

```python
data_city = data_city.drop(data_city[data_city[indicador] <= 0].index)
data_city
```

### Definindo escala:

new_week_cases/100.000 - x*0.01

new_week_deaths/100.000 - x*0.1

incidence_cases - x*0.01

incidence_deaths - x*0.5

cases_growth_factor - x*0.35

deaths_growth_factor - x*2


```python
maximo = max(data_city[indicador])
maximo

escala = {'new_week_cases': 0.01,
          'new_week_deaths': 0.1,
          'incidence_cases': 0.01,
          'incidence_deaths': 0.5,
          'cases_growth_factor': 0.35,
          'deaths_growth_factor': 2}

print(escala[indicador])
```

### Definindo cores:

new_week_cases - Laranja

new_week_deaths - Vermelho

incidence_cases - Laranja

incidence_deaths - Vermelho

cases_growth_factor - Laranja

deaths_growth_factor - Vermelho


```python
cores = {'new_week_cases': '#F5B041',
          'new_week_deaths': '#e64c4e',
          'incidence_cases': '#F5B041',
          'incidence_deaths': '#e64c4e',
          'cases_growth_factor':'#F5B041',
          'deaths_growth_factor': '#e64c4e'}

# Define the color
data_city['fillColor'] = cores[indicador]

data_city
```

### Convertendo dados para json:

```python
def create_geojson_features(data_city):
    features = []

    for _, row in data_city.iterrows():
              
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point',
                'coordinates':[row['longitude'],row['latitude']]
                
            },
            'properties': {
                'time': pd.to_datetime(row['week_last_day'], unit='h').__str__(),
                'style': {'color' : ''},
                'icon': 'circle',
                'popup': '<h5><b>'+titulos[indicador]+':<br><br></h5>'+
                         '<h3><center>'+str(round(row[indicador], 2))+'</b></h3><br></center>'+
                         '<b>Município</b>: '+str(row["city"])+'<br>'+
                         '<b>População</b>: '+str(int(round(row["total_inhabitants"], 0)))+'<br>'+
                         '<b>Estado</b>: '+str(row["state"])+'<br>'+
                         '<b>Região</b>: '+str(row["region"])+'<br>'+
                         '<b>Semana Epidemiológica</b>: '+str(row["epidemiological_week"])+'<br>',
                'iconstyle':{
                    'fillColor': row['fillColor'],
                    'fillOpacity': 0.8,
                    'stroke': 'true',
                    'radius': (row[indicador]*escala[indicador]) + 3,
                }
            }
        }
        features.append(feature)
    return features

start_geojson = create_geojson_features(data_city) 
start_geojson[0]
```

### Plotando com o Folium:

```python
from folium.plugins import TimestampedGeoJson

covid_map = folium.Map(location = [-15.7801, -47.9292],
                    tiles = "openstreetmap",
                    zoom_start = 5)
folium.GeoJson(
    shape_br.set_index('state').to_json(),
    style_function=lambda x: style,
    ).add_to(covid_map)

TimestampedGeoJson(start_geojson,
                  period = 'P1W',
                  duration = 'P1W',
                  transition_time = 1000,
                  auto_play = False).add_to(covid_map)

covid_map.save(indicador+'.html')
```
